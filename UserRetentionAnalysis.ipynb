### Data Cleansing

```python
df_clean = df.copy()

# Convert the oder_date column into datetime
df_clean['order_date'] = df_clean['order_date'].astype('datetime64[ns]')

# create year_month column 
df_clean['year_month'] = df_clean['order_date'].dt.to_period('M')

# Remove all rows without customer_id
df_clean = df_clean[~df_clean['customer_id'].isna()]

# Remove all rows without product_name
df_clean = df_clean[~df_clean['product_name'].isna()]

# Make all product_name lower case
df_clean['product_name'] = df_clean['product_name'].str.lower()

# Remove all lines with product_code or product_name test
df_clean = df_clean[(~df_clean['product_code'].str.lower().str.contains('test')) |
                    (~df_clean['product_name'].str.contains('test '))]
```

```python
# Creating order_status column with the value 'cancelled'. If order_id is started with letter 'c'
# and 'delivered' if the order_id is started without first letter 'c'
df_clean['order_status'] = np.where(df_clean['order_id'].str[:1]=='C', 'cancelled', 'delivered')

# Change the quantity value with negative into positive 
# because the negative value only marked the order as cancelled
df_clean['quantity'] = df_clean['quantity'].abs()

# Remove all lines with the negative price value
df_clean = df_clean[df_clean['price']>0]

# Creating amount of value which multiplication between quantity and price
df_clean['amount'] = df_clean['quantity'] * df_clean['price']

# Change product_name from product_code that has some product_name 
# with one of product_name that is always appeared
most_freq_product_name = df_clean.groupby(['product_code','product_name'], as_index=False).agg(order_cnt=('order_id','nunique')).sort_values(['product_code','order_cnt'], ascending=[True,False])

# Add Ranking number
most_freq_product_name['rank'] = most_freq_product_name.groupby('product_code')['order_cnt'].rank(method='first', ascending=False)

# Filter by rank then drop column order per count and rank column
most_freq_product_name = most_freq_product_name[most_freq_product_name['rank']==1].drop(columns=['order_cnt', 'rank'])

# Merge the new column for most frequent product
df_clean = df_clean.merge(most_freq_product_name.rename(columns={'product_name': 'most_freq_product_name'}), how='left', on='product_code')

# Assign the value of most frequent product to product name
df_clean['product_name'] = df_clean['most_freq_product_name']

# Drop the most frequent product column
df_clean = df_clean.drop(columns='most_freq_product_name')


# Convert customer_id into string 
df_clean['customer_id'] = df_clean['customer_id'].astype(str)
```

```python
# Remove the outlier
from scipy import stats
df_clean = df_clean[(np.abs(stats.zscore(df_clean[['quantity', 'amount']]))<3).all(axis=1)]
df_ckean = df_clean.reset_index(drop=True)
df_clean
```

### Create User Retention Cohort

#### Aggregate transaction data into total summary of transaction/order for each user per month

```python
df_user_monthly = df_clean.groupby(['customer_id','year_month'], as_index=False).agg(order_cnt=('order_id','nunique'))
df_user_monthly
```

#### Create column as cohort from user, for instance cohort first month of transaction

```python
# Assign column cohort with the data from user monthly filter by customer id and year from first month
df_user_monthly['cohort'] = df_user_monthly.groupby('customer_id')['year_month'].transform('min')
df_user_monthly
```

#### Calculate the difference between transaction month with first month of transaction and total it with 1 so the distance of month 0 becomes 1. It means month 0 is first month.

```python
from operator import attrgetter

df_user_monthly['period_num'] = (df_user_monthly['year_month'] - df_user_monthly['cohort']).apply(attrgetter('n')) + 1
df_user_monthly
```

#### Pivot Table with index which is cohort, column is distance of the month and the value is how many unique user (unique count from user id)

```python
df_cohort_pivot = pd.pivot_table(df_user_monthly, index='cohort', columns='period_num', values='customer_id', aggfunc=pd.Series.nunique)
df_cohort_pivot
```

#### Calculate how many users in each cohort (the users that do first time transaction on that month) and divide all value in the pivot table with that value as retention rate

$ \frac{first \; time \; transaction}{total} $

```python
# iloc to extract row from first row
cohort_size  = df_cohort_pivot.iloc[:, 0]
cohort_size
```

```python
df_retention_cohort = df_cohort_pivot.divide(cohort_size, axis=0)
df_retention_cohort
```

#### Show the pivot table that is filled with retention rate value into heatmap

```python
# pip install matplotlib
# pip install seaborn
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns

with sns.axes_style('white'):

    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios':[1, 11]})

    # user retention cohort
    sns.heatmap(df_retention_cohort, annot=True, fmt='.0%', cmap='RdYlGn', ax=ax[1])
    ax[1].set_title('User Retention Cohort')
    ax[1].set(xlabel='Month Number', ylabel='')

    # cohort size
    df_cohort_size = pd.DataFrame(cohort_size)
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(df_cohort_size, annot=True, cbar=False, fmt='g', cmap=white_cmap, ax=ax[0])
    ax[0].tick_params(bottom=False)
    ax[0].set(xlabel='Cohort Size', ylabel='First Order Month', xticklabels=[])

    fig.tight_layout()
```
